# Separating Categorical and Numerical Columns

```python
def separate(loan_data):
    categorical=[]
    numerical=[]
    for columns in loan_data.columns:
        if loan_data[columns].nunique() < 40:
            categorical.append(columns)
        else:
            numerical.append(columns)
    return categorical, numerical

categorical, numerical=separate(loan_data)

from tabulate import tabulate
table=[categorical, numerical]
print(tabulate({"Categorical Columns":categorical,
               "numerical Columns":numerical}, headers=['categorical','numerical']))
```

# User defined functions for level 1 analysis of categorical columns

```python
def fun_cat(col):
    if loan_data[col].isnull().sum()>0:
        loan_data[col].fillna(loan_data[col].mode()[0], inplace=True)
        print(f"\n There are null values in {col} columns")
    print(f"{col}: Mode is {loan_data[col].mode()[0]}")
    print(f"{col}: Number of Missing values in {col} is {loan_data[col].isnull().sum()}")
```

# Does the number of children impact loan defaults?

```python
filtered_data = loan_data[loan_data['CNT_CHILDREN'] <= 5]
default_rates = filtered_data.groupby('CNT_CHILDREN')['TARGET'].mean().reset_index()

plt.figure(figsize=(10, 6))
ax=sb.barplot(data=default_rates, x='CNT_CHILDREN', y='TARGET')
for bar in ax.patches:
        percentage=f"{round(bar.get_height() , 2)}"
        
        x=bar.get_x() + bar.get_width()/2
        y=bar.get_height()
        ax.annotate(percentage, (x,y), va='bottom', ha='center')
plt.title('Loan default rate by number of children')
plt.xlabel('Number of Children')
plt.ylabel('Loan Default Rate')
plt.show()
```

# Observations:
Clients with four children have the highest loan default rate (0.15), while those with 0, 1, 2, or 3 children have similar lower default rates (around 0.08 to 0.09), suggesting that having more children may increase the likelihood of loan default.
